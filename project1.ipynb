```python
df=pd.read_csv("E:\Works\Mojtame Fanni\\numpy_project\Machin learning\Data\\weather.csv")
df.head(8)
```

```python
df.dtypes
```

```python
df.replace({"Yes":1,"No":0},inplace=True)
df.head(5)
```

```python
df.columns
```

```python
df.drop_duplicates(inplace=True)
df.drop(columns=["WindGustDir","WindDir9am","WindDir3pm"],inplace=True)
df.head(5)
```

```python
df.isnull().sum()
```

```python
df1=df.apply(pd.to_numeric, errors='coerce')
df.dropna(inplace=True)
df.shape
```

```python
x=df.iloc[:,:-3].values
y=df.iloc[:,-3].values
x
```

```python
from sklearn.preprocessing import scale,StandardScaler
x=scale(x)
x
```

```python
from sklearn.model_selection import train_test_split
from sklearn import svm
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
Svm=svm.SVC(kernel="linear")
Svm.fit(x_train,y_train)
y_pred_svm=Svm.predict(x_test)
print(f"train_svm_score={Svm.score(x_train,y_train)}")
print(f"test_svm_score={Svm.score(x_test,y_test)}")
```

```python
from sklearn.metrics import confusion_matrix,classification_report
import seaborn as sb
cm_svm=confusion_matrix(y_test,y_pred_svm)
print(classification_report(y_test,y_pred_svm))
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
sb.heatmap(cm_svm, annot=True, fmt="1f",linewidth=3,linecolor="white",cmap="magma",cbar=True)
plt.xlabel("predict")
plt.ylabel("actual")
plt.legend()
plt.title("With SVM")
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
knn_test=KNeighborsClassifier()
parameter={'n_neighbors': np.arange(2, 30, 1)}
knn_cv=GridSearchCV(knn_test, param_grid=parameter, cv=5, verbose=1)
knn_cv.fit(x,y)
```

```python
knn=KNeighborsClassifier(n_neighbors=17)
knn.fit(x_train,y_train)
y_pred_knn=knn.predict(x_test)
print(f"train_knn_score={knn.score(x_train,y_train)}")
print(f"test_knn_score={knn.score(x_test,y_test)}")
```

```python
cm_knn=confusion_matrix(y_test,y_pred_knn)
print(classification_report(y_test,y_pred_knn))
```
